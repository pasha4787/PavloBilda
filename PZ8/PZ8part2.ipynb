## Основи вибіркового методу

Варіант **2** – вибірка: `[1, 7, 7, 10, 5]`.


```python
import math, statistics
import matplotlib.pyplot as plt
from collections import Counter
data = [1,7,7,10,5]
data
```

### Варіаційний ряд, частоти, основні характеристики

```python
n = len(data)
sorted_data = sorted(data)
cnt = Counter(sorted_data)
freq_table = [(x, cnt[x], cnt[x]/n) for x in sorted(cnt)]
x_min, x_max = min(data), max(data)
R = x_max - x_min
mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)
s2 = statistics.variance(data)
s = math.sqrt(s2)
cv = s/mean if mean!=0 else float('nan')
n, sorted_data, freq_table, x_min, x_max, R, mean, median, mode, s2, s, cv
```

### Гістограма

```python
plt.figure()
plt.hist(data, bins=range(min(data), max(data)+2), edgecolor='black')
plt.xlabel('x'); plt.ylabel('Частота'); plt.title('Гістограма вибірки');
plt.show()
```

### Емпірична функція розподілу

```python
xs = sorted_data
Fs = [(x, (i+1)/n) for i,x in enumerate(xs)]
Fs
```

```python
plt.figure()
xs_plot = [xs[0]-0.5] + xs + [xs[-1]+0.5]
F_vals = [0.0] + [i/n for i in range(1,n+1)] + [1.0]
plt.step(xs_plot, F_vals, where='post')
plt.xlabel('x'); plt.ylabel('F*(x)'); plt.title('Емпірична функція розподілу');
plt.ylim(-0.05,1.05)
plt.show()
```

### Box-and-Whisker Plot

```python
plt.figure()
plt.boxplot(data, vert=True, showmeans=True)
plt.ylabel('x'); plt.title('Box-and-Whisker Plot');
plt.show()
```

### Задача 3
N людей за круглим столом. Ймовірність, що A і B сидять поруч.

```python
def p_adjacent_circle(N:int)->float:
    import math
    fav = 2*math.factorial(N-2)
    total = math.factorial(N-1)
    return fav/total

[(N, p_adjacent_circle(N), 2/(N-1)) for N in range(3,11)]
```

### Задача 4
Тризначне число без нулів. Ймовірність, що рівно дві цифри однакові.

```python
digits = range(1,10)
total = 9*9*9
favorable = 0
seen = set()
for a in digits:
  for b in digits:
    if b==a:
      continue
    nums = {100*a+10*a+b, 100*a+10*b+a, 100*b+10*a+a}
    for num in nums:
        if num not in seen:
            seen.add(num)
            favorable += 1
P = favorable/total
total, favorable, P
```

### Задача 5 – «Спортлото 6 з 49»
Ймовірності: 6 вгаданих; 5 або 6; хоча б 1; рівно 2; не менше 4.

```python
from math import comb
total = comb(49,6)
def P_match(k):
    return comb(6,k)*comb(43,6-k)/total

P6 = P_match(6)
P5 = P_match(5)
P4 = P_match(4)
P3 = P_match(3)
P2 = P_match(2)
P1 = P_match(1)
P0 = P_match(0)
{
 'P6': P6,
 'P5or6': P5+P6,
 'Patleast1': 1-P0,
 'P2': P2,
 'P>=4': P4+P5+P6,
 'sum': sum(P_match(k) for k in range(7))
}
```

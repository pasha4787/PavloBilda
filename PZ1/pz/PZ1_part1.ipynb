## Елементи комбінаторики

Варіант: номер студента **n = 2** → задачі **2–6**.


```python
import math

def C(n, k):
    return math.comb(n, k)
```

### Задача 2
На шаховій дошці 8×8:
1) будь-які 2 клітинки;
2) 2 клітинки одного кольору;
3) 2 клітинки різного кольору.

```python
N = 8*8
total_pairs = C(N, 2)
white = black = N//2
same_color = C(white, 2) + C(black, 2)
diff_color = total_pairs - same_color
total_pairs, same_color, diff_color
```

### Задача 3
З цифр 1–5 складаються числа довжиною не більш ніж 3 цифри.
1) без повторень;
2) з повтореннями.

```python
digits = 5
one = digits
two = digits*(digits-1)
three = digits*(digits-1)*(digits-2)
total_no_rep = one + two + three

one_r = digits
two_r = digits**2
three_r = digits**3
total_with_rep = one_r + two_r + three_r
total_no_rep, total_with_rep
```

### Задача 4
У групі 9 людей. Скільки різних підгруп можна створити, якщо в підгрупі має бути не менше 2 людей?

```python
n = 9
all_subsets = 2**n
subgroups_at_least_2 = all_subsets - 1 - n
subgroups_at_least_2
```

### Задача 5
7 різних книг на полиці.
1) дві певні книги стоять поруч;
2) ці дві книги не стоять поруч.

```python
n_books = 7
total_perm = math.factorial(n_books)
block_perm = 2*math.factorial(n_books-1)
not_adjacent = total_perm - block_perm
total_perm, block_perm, not_adjacent
```

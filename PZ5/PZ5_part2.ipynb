## Закони розподілу дискретних випадкових величин

Варіант **2** → задачі **2–6**.


```python
import math
from math import comb
from collections import Counter
import matplotlib.pyplot as plt
```

### Задача 2 – стрілянина до першого влучення, максимум 4 постріли
p=0.7, q=0.3, X – кількість промахів (0–4).

```python
p=0.7
q=1-p
xs=list(range(5))
probs=[]
for k in xs:
    if k<4:
        probs.append(q**k*p)
    else:
        probs.append(q**4)
sum_probs=sum(probs)
EX=sum(x*p_ for x,p_ in zip(xs,probs))
EX2=sum((x**2)*p_ for x,p_ in zip(xs,probs))
VarX=EX2-EX**2
xs, probs, sum_probs, EX, VarX
```

```python
plt.figure()
plt.stem(xs, probs, use_line_collection=True)
plt.xlabel('x'); plt.ylabel('P(X=x)'); plt.title('Розподіл X (Задача 2)'); plt.show()
```

### Задача 3 – X = різниця очок першого і другого кидка

```python
vals=[]
for a in range(1,7):
  for b in range(1,7):
    vals.append(a-b)
cnt=Counter(vals)
xs=sorted(cnt.keys())
probs=[cnt[x]/36 for x in xs]
EX=sum(x*p_ for x,p_ in zip(xs,probs))
EX2=sum((x**2)*p_ for x,p_ in zip(xs,probs))
VarX=EX2-EX**2
xs, probs, EX, VarX
```

```python
plt.figure()
plt.stem(xs, probs, use_line_collection=True)
plt.xlabel('x'); plt.ylabel('P(X=x)'); plt.title('Розподіл X=ξ1-ξ2'); plt.show()
```

### Задача 4 – урна (4 білі, 3 чорні), X – кількість білих у вибірці з 3

```python
white, black = 4,3
total = comb(white+black,3)
xs=[0,1,2,3]
probs=[]
for k in xs:
    if 0<=k<=3:
        probs.append(comb(white,k)*comb(black,3-k)/total)
    else:
        probs.append(0)
sum_probs=sum(probs)
EX=sum(x*p_ for x,p_ in zip(xs,probs))
EX2=sum((x**2)*p_ for x,p_ in zip(xs,probs))
VarX=EX2-EX**2
xs, probs, sum_probs, EX, VarX
```

```python
plt.figure()
plt.stem(xs, probs, use_line_collection=True)
plt.xlabel('x'); plt.ylabel('P(X=x)'); plt.title('Розподіл X (Задача 4)'); plt.show()
```

### Задача 6 – два стрілки, X – кількість влучень

```python
p1,p2=0.5,0.4
q1,q2=1-p1,1-p2
xs=[0,1,2]
p0=q1*q2
p2_=p1*p2
p1_=1-p0-p2_
probs=[p0,p1_,p2_]
EX=sum(x*p_ for x,p_ in zip(xs,probs))
EX2=sum((x**2)*p_ for x,p_ in zip(xs,probs))
VarX=EX2-EX**2
xs, probs, EX, VarX
```

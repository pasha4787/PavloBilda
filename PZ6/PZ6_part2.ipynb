### Задача 2 – сума двох U(0,2)
X,Y ~ U(0,2), Z=X+Y. Трикутний розподіл на [0,4].

```python
def f_Z(z:float)->float:
    if 0<z<=2:
        return z/4
    elif 2<z<4:
        return (4-z)/4
    else:
        return 0.0

def F_Z(z:float)->float:
    if z<=0:
        return 0.0
    elif z<=2:
        return z**2/8
    elif z<4:
        return 1-(4-z)**2/8
    else:
        return 1.0
[ (x, F_Z(x)) for x in [0,1,2,3,4] ]
```

### Задача 3 – максимум з експоненційних (λ=10)

```python
def F_Z_max(z:float,n:int,lam:float=10.0)->float:
    if z<0:
        return 0.0
    Fx=1-math.exp(-lam*z)
    return Fx**n
[ (x, F_Z_max(x,5)) for x in [0,0.1,0.2] ]
```

### Задача 4 – Z = sin X, X~U(0,π)

```python
def sample_Z_sin(N=50000):
    vals=[]
    for _ in range(N):
        x = random.random()*math.pi
        vals.append(math.sin(x))
    m=sum(vals)/N
    var=sum((v-m)**2 for v in vals)/N
    return m,var
sample_Z_sin(20000)
```

### Задача 5 – Z = cos X, X~U(0,π)

```python
def sample_Z_cos(N=50000):
    vals=[]
    for _ in range(N):
        x = random.random()*math.pi
        vals.append(math.cos(x))
    m=sum(vals)/N
    var=sum((v-m)**2 for v in vals)/N
    return m,var
sample_Z_cos(20000)
```

### Задача 6 – максимум з експоненційних (λ=5)

```python
def F_Z_max_exp(z:float,n:int,lam:float=5.0)->float:
    if z<0:
        return 0.0
    Fx=1-math.exp(-lam*z)
    return Fx**n
[ (x, F_Z_max_exp(x,3)) for x in [0,0.1,0.2] ]
```

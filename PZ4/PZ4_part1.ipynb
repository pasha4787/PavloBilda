### Задача 2
Найбільш ймовірне число влучень при 5 пострілах, p=0.8.

```python
n = 5
p = 0.8
q = 1-p
probs = []
for k in range(n+1):
    pk = math.comb(n,k)*(p**k)*(q**(n-k))
    probs.append((k, pk))
probs, max(probs, key=lambda t:t[1])
```

### Задача 3
Найбільш ймовірне число гербів при 25 киданнях монети.

```python
n = 25
p = 0.5
q = 1-p
probs = [(k, math.comb(n,k)*(p**k)*(q**(n-k))) for k in range(n+1)]
max(probs, key=lambda t:t[1])
```

### Задача 4
Монету кинуто 10 разів. P(4≤X≤6), P(X≥1).

```python
def binom_p(n,k,p):
    return math.comb(n,k)*(p**k)*((1-p)**(n-k))
n=10;p=0.5
P_4_6 = sum(binom_p(n,k,p) for k in range(4,7))
P_ge1 = 1-binom_p(n,0,p)
P_4_6, P_ge1
```

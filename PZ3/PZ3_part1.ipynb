### Задача 3
Автобус (2 хв), пішохід (15 хв), інтервал 25 хв. Ймовірність, що автобус наздожене пішохода.

```python
U = sp.symbols('U', real=True)
vp = 1/15
vb = 1/2
t = (vp*U - vb*25)/(vp-vb)
t_simpl = sp.simplify(t)
cond = sp.solve_univariate_inequality(t_simpl <= U+15, U)
t_simpl, cond
```

### Задача 5 (задача про зустріч)
Дві людини приходять між 0 і 60 хв, чекають по 20 хв. Ймовірність зустрічі.

```python
side = 60
total_area = side*side
tri_area = (side-20)**2/2
bad_area = 2*tri_area
good_area = total_area - bad_area
P5 = good_area/total_area
P5
```
